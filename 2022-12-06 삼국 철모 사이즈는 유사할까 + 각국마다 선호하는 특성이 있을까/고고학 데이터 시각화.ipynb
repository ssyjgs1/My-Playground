```python
import numpy as np
# 데이터 설정
categories = ['남성+A', '남성+B1', '남성+B2', '남성+C', '여성+B1', '여성+B2', '여성+C']
quantity = [1, 7, 7, 1, 1, 4, 2]

# 데이터를 2차원 배열로 변환
data = np.array(quantity).reshape(1, -1)

# 히트맵 그리기
plt.imshow(data, cmap='Blues')

# 축 설정
plt.xticks(np.arange(len(categories)), categories)
plt.yticks([])  # y축 레이블 없음

# 컬러바 추가
cbar = plt.colorbar()
cbar.set_label('Quantity')

# 그래프 출력
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 설정
categories = ['A', 'B1', 'B2', 'C']
genders = ['남성', '여성']
data = [
    [1, 0, 3, 0],
    [2, np.nan, 1, 1]
]

# 데이터 전처리
data = np.array(data)

# 히트맵 그리기
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='Blues')

# 축 설정
ax.set_xticks(np.arange(len(categories)))
ax.set_yticks(np.arange(len(genders)))
ax.set_xticklabels(categories)
ax.set_yticklabels(genders)

# 텍스트 주석 추가
for i in range(len(genders)):
    for j in range(len(categories)):
        if not np.isnan(data[i, j]):
            text = ax.text(j, i, int(data[i, j]), ha='center', va='center', color='black')
        else:
            text = ax.text(j, i, '?', ha='center', va='center', color='black')

# 컬러바 추가
cbar = plt.colorbar(im)
cbar.set_label('Quantity')

# 그래프 출력
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# 데이터 설정
categories = ['A', 'B1', 'B2', 'C']
genders = ['남성', '여성']
data = [
    [1, 0, 3, 0],
    [2, np.nan, 1, 1]
]

# 데이터 전처리
data = np.array(data)

# 컬러맵 설정
colors = [(0, 'black'), (0.3, 'brown'), (0.6, 'gold'), (1, 'yellow')]
cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

# 히트맵 그리기
fig, ax = plt.subplots()
im = ax.imshow(data, cmap=cmap)

# 배경색 설정
ax.set_facecolor('black')

# 축 설정
ax.set_xticks(np.arange(len(categories)))
ax.set_yticks(np.arange(len(genders)))
ax.set_xticklabels(categories)
ax.set_yticklabels(genders)

# 텍스트 주석 추가
for i in range(len(genders)):
    for j in range(len(categories)):
        if not np.isnan(data[i, j]):
            text = ax.text(j, i, int(data[i, j]), ha='center', va='center', color='white')
        else:
            text = ax.text(j, i, '?', ha='center', va='center', color='white')

# 컬러바 추가
cbar = plt.colorbar(im)
cbar.set_label('Quantity')

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 설정
categories = ['A', 'B1', 'B2', 'C']
genders = ['남성', '여성']
data = [
    [1, 1, 2, 1],
    [0, 3, 2, 1],
    [1, 1, 1, 2],
    [1, 1, 0, 2],
    [1, 1, 2, 1],
    [2, 2, 1, 0],
    [1, 0, 1, 2],
    [0, 2, 2, 1]
]

# 데이터 전처리
x = []
y = []
colors = []
for i in range(len(genders)):
    for j in range(len(categories)):
        x.append(j)
        y.append(i)
        colors.append(data[i][j])

# 산점도 그리기
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=colors, cmap='YlOrBr')

# 축 설정
ax.set_xticks(range(len(categories)))
ax.set_yticks(range(len(genders)))
ax.set_xticklabels(categories)
ax.set_yticklabels(genders)

# 컬러바 추가
cbar = plt.colorbar(scatter)
cbar.set_label('Quantity')

# 그래프 출력
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 설정
data = {
    '성별': ['남성', '남성', '남성', '남성', '여성', '남성', '남성', '남성', '남성', '남성', '남성', '남성', '여성', '여성', '남성', '남성', '여성', '남성', '여성', '남성', '남성', '여성', '여성'],
    '무장조합': ['A', 'B2', 'B1', 'B2', 'B2', 'B2', 'B1', 'B1', 'B2', 'C', 'B1', 'B1', 'C', 'B2', 'B1', 'B2', 'C', 'B1', 'B2', 'B2', 'B2', 'B1', 'B2']
}

df = pd.DataFrame(data)

# 데이터 변환
pivot_table = df.pivot_table(index='성별', columns='무장조합', aggfunc=len, fill_value=0)

# 히트맵 그리기
sns.heatmap(data=pivot_table, cmap='Blues', annot=True, fmt='d')

# 그래프 출력
plt.show()
```
